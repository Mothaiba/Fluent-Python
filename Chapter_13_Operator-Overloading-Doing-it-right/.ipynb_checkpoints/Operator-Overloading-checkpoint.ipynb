An unary or infix operator (e.g. -, ~, +, *) should never change its operands.

Only augmented operators (e.g. +=) should change their first operand.

When processing a + b, Python tries a.\_\_add\_\_ first, if this method could not provide the desired result, Python will try b.\_\_radd\_\_

#### Rich Comparision Operators

Note from the above:
- == will always return a value, because it eventually fallback to comparing the id of the operands.
- != will fallback to use == if \_\_ne\_\_ methods cannot handle, thus it means != will also always return a value. Furthermore, in almost all cases, we do not need to implement \_\_ne\_\_ at all.
- Ordering operators fallback to their opposites (less than => greater than or equal to, etc.)

When overloading operands, it is recommended to return NotImplement (without Error) if we do not want to handle the cases. This will direct Python to try to find another way to complete the computation instead of directly throwing an Error.

#### Augmented Assignment Operators

\_\_iadd\_\_ and \_\_imul\_\_ will fallback to \_\_add\_\_ and \_\_mul\_\_ if not defined.

For immutable types, we should not implement augmented operators, just let them fallback to infix operators is ok.
